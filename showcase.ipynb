```python
import librosa
import os
import soundfile as sf
import IPython.display as ipd
```

```python
filename_fn = "/content/drive/MyDrive/CS 1699 Final Project/Project/showcase/Female_Neutral.wav"
filename_fh = "/content/drive/MyDrive/CS 1699 Final Project/Project/showcase/Female_Happy.wav"
filename_ff = "/content/drive/MyDrive/CS 1699 Final Project/Project/showcase/Female_Fearful.wav"

filename_mn = "/content/drive/MyDrive/CS 1699 Final Project/Project/showcase/Male_Neutral.wav"
filename_mh = "/content/drive/MyDrive/CS 1699 Final Project/Project/showcase/Male_Happy.wav"
filename_mf = "/content/drive/MyDrive/CS 1699 Final Project/Project/showcase/Male_Fearful.wav"
```

```python
# Female_Neutral
time_series, sample_rate = librosa.load(filename_fn, sr=None)
ipd.Audio(time_series, rate = sample_rate)
```

```python
# Female_Neutral -> 8khz
EIGHT_KHZ = 8192
resampled_time_series = librosa.core.resample(y=time_series, orig_sr=sample_rate, target_sr=EIGHT_KHZ)
ipd.Audio(resampled_time_series, rate = EIGHT_KHZ)
```

```python
# Female_Neutral -> Phone calling
SPEECH_LOW_BAND = 200
SPEECH_UPPER_BAND = 3400
short_time_fourier_transform = librosa.core.stft(y=resampled_time_series)
short_time_fourier_transform[:SPEECH_LOW_BAND] = 0
short_time_fourier_transform[SPEECH_UPPER_BAND: len(short_time_fourier_transform)] = 0

reconstructed_time_series = librosa.core.istft(short_time_fourier_transform)
ipd.Audio(reconstructed_time_series, rate = EIGHT_KHZ)
```
